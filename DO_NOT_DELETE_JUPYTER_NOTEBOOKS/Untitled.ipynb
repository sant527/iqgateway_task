```python
import requests
from project_settings.pretty_printing import dumps, jprint
import json

payload = {
    "username": "u000",
    "password": "123"
}
#url='http://webapp:8000/dj-rest-auth/login/'
url='http://host.docker.internal:8028/dj-rest-auth/login/'
# **dont use Accept plain/Text" in header it will fail
headers = {
    "Content-Type" : "application/json",
}

try:
    r = requests.post(
        url,
        data=json.dumps(payload),
        headers=headers
    )
    r.raise_for_status()
    print(jprint(r.json()))
except requests.exceptions.HTTPError as errh:
    print ("Http Error:",errh)
    print(jprint(errh.response.json()))
except requests.exceptions.ConnectionError as errc:
    print ("Error Connecting:",errc)
except requests.exceptions.Timeout as errt:
    print ("Timeout Error:",errt)
except requests.exceptions.RequestException as err:
    print ("OOps: Something Else",err)
```
